```python
class D1(nn.Module):
  """"
  parent discriminator

  implement many to one lstm
  that takes sequence and gives a single output

  """

  def __init__(self):
      super(D1, self).__init__()

      self.input_size = 16
      self.hidden_size = self.input_size
      self.num_layers = 2

      self.lstm1 = nn.LSTM(input_size=self.input_size, hidden_size=self.hidden_size, num_layers=self.num_layers, batch_first = True, bidirectional=True)
      self.linear1 = nn.Linear(in_features=2* self.hidden_size, out_features=1)
      self.a1 = nn.Sigmoid()


  def forward(self,sequence_input):
#     input: tensor of shape ( N, L, H_{in})
      opt1,(ho,co) = self.lstm1(sequence_input)
      opt2 = self.linear1(opt1)
      opt3 = self.a1(opt2)
      return opt3, (ho,co)

  def init_hidden(self, batch_size):
#     h_0: tensor of shape (D *{num_layers}, N, H_{out})
      ho = torch.random(2 * self.num_layers, batch_size, self.hidden_size )
#     ​c_0: tensor of shape (D *{num_layers}, N, H_{cell})
      co = torch.random(2 * self.num_layers, batch_size, self.hidden_size )
      return (ho,co)
```

```python
class G1(nn.Module):
  """"
  parent generator
  implement one to many lstm
  that takes single input and gives a sequence output
  """

  def __init__(self):
      super(G1, self).__init__()
      self.input_size = 16
      self.hidden_size = self.input_size 
      self.num_layers = 2
      self.noise_size = 10
      self.batch_size = 8

    
      self.linear1 = nn.Linear(in_features=self.noise_size, out_features= self.batch_size * self.input_size  * self.hidden_size)
      self.lstm1 = nn.LSTM(input_size=self.hidden_size, hidden_size=self.hidden_size, num_layers=self.num_layers, batch_first = True)
      self.a1 = nn.Tanh()


  def forward(self):
#     input: tensor of shape ( N, L, H_{in})
      noise = torch.randn(self.noise_size)
      opt1 = self.linear1(noise)
      opt1 = opt1.reshape(self.batch_size ,self.input_size  , self.hidden_size)
      opt2,(ho,co) = self.lstm1(opt1)
      opt3 = self.a1(opt2)
      return opt3, (ho,co)

  def init_hidden(self, batch_size):
#     h_0: tensor of shape (D *{num_layers}, N, H_{out})
      ho = torch.random(self.num_layers, batch_size, self.hidden_size )
#     ​c_0: tensor of shape (D *{num_layers}, N, H_{cell})
      co = torch.random(self.num_layers, batch_size, self.hidden_size )
      return (ho,co)

      
```

```python
# # apply auto encoder
class G2(nn.Module):
  """"
  child generator
  ( autoencoder using LSTM )
  implement many to many lstm
  that takes sequence and gives a single output
  """

  def __init__(self):
      super(G2, self).__init__()
      self.input_size = 16
      self.hidden_size = self.input_size 
      self.num_layers = 2
      self.noise_size = 10
      self.batch_size = 8

      self.lstm1 = nn.LSTM(input_size=self.hidden_size, hidden_size=self.hidden_size, num_layers=self.num_layers, batch_first = True)    
      self.linear1 = nn.Linear(in_features=self.batch_size * self.input_size  * self.hidden_size , out_features= self.noise_size)
      self.linear2 = nn.Linear(in_features=self.noise_size, out_features= self.batch_size * self.input_size  * self.hidden_size)
      self.lstm2 = nn.LSTM(input_size=self.hidden_size, hidden_size=self.hidden_size, num_layers=self.num_layers, batch_first = True)
      self.a1 = nn.Tanh()


  def forward(self, sequence_input):
#     input: tensor of shape ( N, L, H_{in})
      opt1,(ho,co) = self.lstm1(sequence_input)
      opt1 = opt1.reshape(self.batch_size * self.input_size  * self.hidden_size)
      opt2 = self.linear1(opt1)
      opt3 = self.linear2(opt2)
      opt4 = opt3.reshape(self.batch_size ,self.input_size  , self.hidden_size)
      opt5,(ho,co) = self.lstm2(opt4)
      opt6 = self.a1(opt5)
      return opt6, (ho,co)

  def init_hidden(self, batch_size):
#     h_0: tensor of shape (D *{num_layers}, N, H_{out})
      ho = torch.random(self.num_layers, batch_size, self.hidden_size )
#     ​c_0: tensor of shape (D *{num_layers}, N, H_{cell})
      co = torch.random(self.num_layers, batch_size, self.hidden_size )
      return (ho,co)


```

```python
def train_discriminator1(sequence_input):
    # Create the labels which are later used as input for the BCE loss
    real_labels = torch.ones(2* batch_size, 1).to(device)    #torch.Size([16, 1]) 
    fake_labels = torch.zeros(2* batch_size, 1).to(device)   #torch.Size([16, 1]) 
    # Loss for real music sequence
    d1_outputs1,(h0,c0) = d1(sequence_input)
    d1_outputs1 = d1_outputs1[-1]
    d1_loss_real = d1_loss_function(d1_outputs1.to(device) , real_labels)
    d1_real_score = d1_outputs1



    # Loss for fake images
    g1_fake_images,(h0,c0) = g1()
    d1_outputs2,(h0,c0) = d1(g1_fake_images)

    d1_outputs2 = d1_outputs2[-1]
    d1_loss_fake = d1_loss_function(d1_outputs2.to(device), fake_labels)
    d1_fake_score = d1_outputs2

    # Combine losses
    d1_loss = d1_loss_real + d1_loss_fake
    # Reset gradients
    d1_optimizer.zero_grad()
    g1_optimizer.zero_grad()
    # Compute gradients
    d1_loss.backward()
    # Adjust the parameters using backprop
    d1_optimizer.step()
    
    return d1_loss, d1_real_score, d1_fake_score
```

```python
def train_discriminator2(sequence_input, parent_loss):
    # Create the labels which are later used as input for the BCE loss
    real_labels = torch.ones(2*batch_size, 1).to(device)
    fake_labels = torch.zeros(2*batch_size, 1).to(device)
    # Loss for real images
    d2_outputs1,(h0,c0) = d2(sequence_input)
    d2_outputs1 = d2_outputs1[-1]
    d2_loss_real = d2_loss_function(d2_outputs1.to(device), real_labels)
    d2_real_score = d2_outputs1
    # Loss for fake images
    z2,(h0,c0) = g1()
    g2_fake_images,(h0,c0) = g2(z2)

    d2_outputs2, (h0,c0) = d2(g2_fake_images)
    d2_outputs2 = d2_outputs2[-1]
    d2_loss_fake = d2_loss_function(d2_outputs2.to(device), fake_labels)
    d2_fake_score = d2_outputs2
    # Combine losses
    d2_loss = d2_loss_real + d2_loss_fake + parent_loss
    # Reset gradients
    d2_optimizer.zero_grad()
    g2_optimizer.zero_grad()
    # Compute gradients
    d2_loss.backward()
    # Adjust the parameters using backprop
    d2_optimizer.step()
    
    return d2_loss, d2_real_score, d2_fake_score
```

```python
def train_generator1():
    # Generate fake images and calculate loss
    g1_fake_images,(h0,c0) = g1()

    labels = torch.ones(2*batch_size, 1).to(device)

    opt,(h0,c0) = d1(g1_fake_images)
    opt = opt[-1]
    g1_loss = d1_loss_function(opt.to(device), labels)

    # Backprop and optimize
    d1_optimizer.zero_grad()
    g1_optimizer.zero_grad()

    with torch.no_grad():
      g1_loss.backward()
      g1_optimizer.step()
    return g1_loss, g1_fake_images
```

```python
def train_generator2(sequence_input ):
    # Generate fake images and calculate loss
    g2_fake_images,(h0,c0) = g2(sequence_input)

    labels = torch.ones(2*batch_size, 1).to(device)


    opt,(h0,c0) = d2(g2_fake_images)
    opt = opt[-1]
    g2_loss = d2_loss_function(opt.to(device), labels)

    # Backprop and optimize
    d2_optimizer.zero_grad()
    g2_optimizer.zero_grad()

    with torch.no_grad():
      g2_loss.backward()
      g2_optimizer.step()
    return g2_loss, g2_fake_images
```
